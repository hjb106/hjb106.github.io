merge 방법 4가지 
1) Inner join  A N B
2) left outer join : 왼쪽 기준 합치기 
3) right outer join : 오른쪽 기준 합치기 
4) full outer join: 양쪽 기준 합치기 A U B

```python
pd.merge(vp, vq)
```

```python
pd.merge(vp, vq, on='Product', how='inner')
```

```python
pd.merge(vp, vq, on='Product', how='left')
```

```python
pd.merge(vp, vq, on='Product', how='right')
```

```python
pd.merge(vp, vq, on='Product', how='outer')
```
